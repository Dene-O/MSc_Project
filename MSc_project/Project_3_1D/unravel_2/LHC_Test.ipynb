```python
from scipy.stats.qmc import LatinHypercube
import matplotlib.pyplot as plt
```

```python
n_features = 2
n_samples  = 100
bounds     = 1.5
F_Ranges   = [10, 4]
F_mid      = [10, 6]
```

```python
sampler = LatinHypercube(d = n_features, strength = 1)
            
sampled_dist = sampler.random(n = n_samples) * bounds - (bounds / 2)

sampled_dist = sampled_dist * F_Ranges + F_mid
```

```python
print(sampled_dist)
```

```python
fig, ax1 = plt.subplots()
        

ax1.scatter(sampled_dist[:,0], sampled_dist[:,1], marker = '.')

plt.show()
                        
```
