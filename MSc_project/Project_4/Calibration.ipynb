```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10) + 1
x = x.reshape([10,1])
y = 10 - (10 / x)
y = x + 3
print(x,y)
```

```python
LR = LinearRegression(fit_intercept=False)
LR.fit(x, y)
```

```python
X= np.array([0]).reshape([1,-1])
lower = LR.predict(X).item()
X=X+10
upper = LR.predict(X).item()
print(lower, upper)
```

```python
fig, ax = plt.subplots()
ax.scatter(x.ravel(),y.ravel())
ax.plot([0,10],[lower,upper])
plt.show()
```

```python
x = x.ravel() / 10
y = y.ravel() / upper
print(x,y)
```

```python
fig, ax = plt.subplots()
ax.scatter(x,y)
ax.plot([0,1],[0,1])
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_true = x, y_pred = y)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_true = x, y_pred = y)
```
